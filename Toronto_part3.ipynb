## Segmenting and Clustering Neighborhoods in Toronto

***Firstly, I have used Beautiful soup package  to parse the web page.***

```python
# import the library we use to open URLs
import urllib.request
# import the BeautifulSoup library so we can parse HTML and XML documents
from bs4 import BeautifulSoup

url = 'https://en.wikipedia.org/wiki/List_of_postal_codes_of_Canada:_M'
page = urllib.request.urlopen(url)

# parse the HTML from our URL into the BeautifulSoup parse tree format
soup = BeautifulSoup(page, "lxml")
```

```python
#Finding the table tag from the html page
right_table=soup.find('table', class_='wikitable')
```

***Assigning the web table data into three different lists and assigning the data to a dataframe using pandas.***

```python
#Organising the data into 3 cloumns
A=[]
B=[]
C=[]


for row in right_table.findAll('tr'):
    cells=row.findAll('td')
    if len(cells)==3:
        A.append(cells[0].find(text=True))
        B.append(cells[1].find(text=True))
        C.append(cells[2].find(text=True))
```

```python
import pandas as pd
import numpy as np

#Assiging the headers and applying pandas to create dataframe
df=pd.DataFrame(A,columns=['Postalcode'])
df['Borough']=B
df['Neighborhood']=C
df
```

***Removing the data in Borough whose values are 'Not assigned' & extra spaces from the dataframe***

```python
df = df.replace('Not assigned',np.nan, regex=True)
df.dropna(subset=["Borough"], axis=0, inplace=True)
df = df.reset_index(drop=True)
df[df.columns] = df.apply(lambda x: x.str.strip())  #Stripping extra space from the columns
df
```

```python
df.shape
```

***I have used CSV file method to get the coordinates,I have performed inner join method on the dataframes to combine the data***

```python
df_geodata = pd.read_csv('http://cocl.us/Geospatial_data')
df_geodata.rename(columns={'Postal Code':'Postalcode'},inplace=True)
df_inner = pd.merge(df, df_geodata, on='Postalcode', how='inner')
df_inner
```

#### *Importing the libraries for the visualisation of maps and clusters* ####

```python

#!conda install -c conda-forge geopy --yes # uncomment this line if you haven't completed the Foursquare API lab
from geopy.geocoders import Nominatim # convert an address into latitude and longitude values

import requests # library to handle requests
from pandas.io.json import json_normalize # tranform JSON file into a pandas dataframe

# Matplotlib and associated plotting modules
import matplotlib.cm as cm
import matplotlib.colors as colors

# import k-means from clustering stage
from sklearn.cluster import KMeans

#!conda install -c conda-forge folium=0.5.0 --yes # uncomment this line if you haven't completed the Foursquare API lab
import folium # map rendering library

print('Libraries imported.')
```

***Assinging the address to Toronto and getting the coordinates of the city***

```python
address = 'Toronto'

geolocator = Nominatim(user_agent="tr_explorer")
location = geolocator.geocode(address)
latitude = location.latitude
longitude = location.longitude
print('The geograpical coordinate of Toronto City  are {}, {}.'.format(latitude, longitude))
```

***Map view with the neighbhooring cities***

```python
# create map of New York using latitude and longitude values
map_toronto = folium.Map(location=[latitude, longitude], zoom_start=10)

# add markers to map
for lat, lng, borough, neighborhood in zip(df_inner['Latitude'], df_inner['Longitude'], df_inner['Borough'], df_inner['Neighborhood']):
    label = '{}, {}'.format(neighborhood, borough)
    label = folium.Popup(label, parse_html=True)
    folium.CircleMarker(
        [lat, lng],
        radius=5,
        popup=label,
        color='blue',
        fill=True,
        fill_color='#3186cc',
        fill_opacity=0.7,
        parse_html=False).add_to(map_toronto)  
    
map_toronto
```
